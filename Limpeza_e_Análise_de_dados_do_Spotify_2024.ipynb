```python
#IMPORTAÇÃO DO ARQUIVO DO DRIVE
df = pd.read_csv('/content/drive/MyDrive/JOBS/SPOTIFY DATA/Most Streamed Spotify Songs 2024.csv',encoding="ISO-8859-1")
```

```python
#DEFINIÇÃO DE QUANTIDADE DE COLUNAS
pd.set_option('display.max_columns',85)
df.head()
```

```python
#VERIFICAÇÃO DOS TIPOS DE DADOS PRESENTES NO DATAFRAME
print(df.info())
```

```python
#REMOVENDO AS COLUNAS 'Soundcloud Streams' E 'TIDAL Popularity' POR FUGIREM DO PADRÃO APRESENTADO DE VALORES NÃO NULOS
df.drop(columns=['Soundcloud Streams','TIDAL Popularity'], axis=1,inplace=True)
df.head()
```

```python
df.head()
```

```python
#APLICAÇÃO DA FUNÇÃO MAP EM TODAS AS COLUNAS DO DATAFRAME E REMOÇÃO DAS VÍRGULAS PRESENTES NAS
# CÉLULAS TROCANDO POR ESPAÇOS
df = df.applymap(lambda x: x.replace(',','') if isinstance(x,str) else x)
```

```python
#CONVERSÃO DE VALORES DO TIPO OBJECT PARA TIPO NUMÉRICO a partir da coluna 06
for x in df.columns[6:]:
    df[x] = pd.to_numeric(df[x])
    print(df[x].dtype)

df.info()
```

```python
#PADRONIZANDO AS CÉLULAS DA COLUNA TRACK
df['Track'] = df['Track'].apply(lambda x: x[:20] + '...' if len(x) > 30 else x)
df
```
